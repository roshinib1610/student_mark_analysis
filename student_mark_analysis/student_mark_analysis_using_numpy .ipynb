```python
import numpy as np

np.random.seed(25)
marks=np.random.randint(10,101,size=(20,5))
print(marks)
```

```python
#1.average by subjects and students

average_by_subjects=np.mean(marks,axis=0)
average_by_students=np.mean(marks,axis=1)

print("average by subjects:",average_by_subjects)
print("average by students:",average_by_students)
```

```python
#2.highest and lowest per subject

highest=np.max(marks,axis=0)
lowest=np.min(marks,axis=0)

print("highest marks of the subject:",highest)
print("lowest marks of the subject:",lowest)
```

```python
#3.overall class topper

total_marks=np.sum(marks,axis=1)
topper=np.argmax(total_marks)

print("overall class topper is index",topper,"with marks of",total_marks[topper])
```

```python
#4.pass count per subject

pass_fail=marks>=40
pass_count=np.sum(pass_fail,axis=0)
print("pass count by subject:",pass_count)
```

```python
#5.which subject is difficult

difficult_subject=np.argmin(average_by_subjects)
print("difficult subject is:",difficult_subject,"with average marks of",average_by_subjects[difficult_subject])
```

```python
#6.ranking students

ranks=np.argsort(-total_marks)
print("first rank of the class",ranks[0])
print("first 5 rank of the class",ranks[:5])
```
